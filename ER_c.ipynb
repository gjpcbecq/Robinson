```python
def CToRobinson(X): 
    [N, LX] = X.shape
    Y = numpy.empty((N * LX))
    for I in range(N): 
        for J in range(LX): 
            IJ = LX * I + J
            Y[IJ] = X[I, J]
    return Y
```
